```python
import pandas as pd
```

```python
# Overview of the data in the SEC_X table
df = pd.read_csv('TB_SEC_X.csv', encoding='cp1252')
df_EDA = pd.DataFrame({
    'column name' : df.columns,
    'dtype' : df.dtypes,
    'number of unique values' : df.nunique(),
    'number of entries' : df.count()
}).reset_index().drop(columns='index')

df_EDA
```

```python
# Create method to print the column name and dtype to use with the SQL Schema
def table_schema(df):
    df_schema = pd.DataFrame({
    'column name' : df.columns,
    'dtype' : df.dtypes
    }).reset_index().drop(columns='index')
    df_schema.loc[(df_schema['dtype'] == 'int64'),'dtype'] = 'INTEGER'
    df_schema.loc[(df_schema['dtype'] == 'float64'),'dtype'] = 'DECIMAL'
    df_schema.loc[(df_schema['dtype'] == 'object'),'dtype'] = 'VARCHAR'
    for index, row in df_schema.iterrows():
        print(row['column name'], row['dtype'], ',')
```

```python
# Row and postgres dtype for the TB_SEC_X table
table_schema(pd.read_csv('TB_SEC_X.csv', encoding='cp1252'))
```

```python
# Row and postgres dtype for the TSDem table
table_schema(pd.read_csv('TSDem.csv', encoding='cp1252'))
```

```python
# Row and postgres dtype for the TVIV table
table_schema(pd.read_csv('TVIV.csv', encoding='cp1252'))
```
