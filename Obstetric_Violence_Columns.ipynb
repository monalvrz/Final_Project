```python
import pandas as pd
```

```python
# List of csv files that will be used
# TB_SEC_III
# TB_SEC_IV
# TB_SEC_X
# TSDem.csv
# TVIV.csv

# Create an array with the filenames that were previously mentioned
file_list = ['TVIV.csv','TSDem.csv','TB_SEC_III.csv','TB_SEC_IV.csv','TB_SEC_X.csv']

# Create an empty dictionary that will be used to store the file name as a key and a list of all its columns as its value 
col_dict = {}

# Create a dataframe after reading each csv and get the column names as a list which is stored as a key : value pair
for file in file_list:
    column_list = pd.read_csv(file,encoding='latin1').columns.values.tolist()
    col_dict[file] = column_list
```

```python
# Create a Dataframe that shows which file has which column
df_columns = pd.DataFrame(columns=['File','Column'])

# Add to each row the column from the csv file and its source csv file
for key,values in col_dict.items():
    for value in values:
        row = {
            'File': key,
            'Column' : value
        }
        df_columns = df_columns.append(row,ignore_index=True)
df_columns.head(10)
```

```python
# Drop the duplicate column names, regardless of the file name
df_columns.drop_duplicates(subset='Column',inplace=True)
# Reset the DataFrame Index
df_columns.reset_index()
# Show the resulting dataframe
df_columns
```

```python
# Loop through the resulting DataFrame and print the contents using the format that can be copy pasted to the schema.sql file to avoid having to type all this manually
for index,row in df_columns.iterrows():
    file_name = row['File'].replace('.csv','')
    print(f'"{file_name}"."{row["Column"]}",')
```
